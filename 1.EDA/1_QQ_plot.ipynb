## QQ Plot

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# Gaussian Dist. N(0,1)
std_normal=np.random.normal(loc=0,scale=1,size=1000)


# 0 to 100th percentile of std-normal
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
import matplotlib.pyplot as plt
```

```python
# generate 100 samples from N(20,5)
measurements=np.random.normal(loc=0,scale=1,size=100)

# generating qq plot
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```
