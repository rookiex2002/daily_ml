## QQ Plot :

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


x=np.random.normal(loc=20,scale=5,size=100)
stats.probplot(x,dist='norm',plot=plt)
plt.show();
```

```python
x=np.random.normal(loc=20,scale=5,size=1000)
stats.probplot(x,dist='norm',plot=plt)
plt.show();
```

```python
# the values are not standardised if we standardise them we get line at 45 degree.
```

### Exponential distribution

```python
x=np.random.exponential(scale=1.0,size=1000)
sns.set()
ax=sns.distplot(x)
plt.show()
```

```python
# plot CDF (X)
kwargs={'cumulative':True}
sns.distplot(x,hist_kws=kwargs,kde_kws=kwargs)
```

```python
# QQ-Plot

stats.probplot(x,dist='norm',plot=plt)
plt.grid()
```

## Box-cox Transform

```python
x_t,l=stats.boxcox(x) # l-lambda , x_t=x transformed by box-cox
print(l)

## QQ plot
stats.probplot(x_t,dist='norm',plot=plt)
plt.grid()

# Expo ----> (~ Gaussian ) using boxcox
```

```python
# if we want to use x_t :we have to standardize (mean-ceiling + var scaling) if data used in model building.
```

```python
# PDF of x_t
sns.set()
ax=sns.distplot(x_t)

# CDF of x_t
kwargs={'cumulative':True}
sns.distplot(x_t,hist_kws=kwargs,kde_kws=kwargs)
```

```python
# Pareto distributed data (most people vs very fewer product)
x=np.random.pareto(a=2.0,size=1000)

# PDF of x
sns.set()
ax=sns.distplot(x)

# CDF of x
kwargs={'cumulative':True}
sns.distplot(x,hist_kws=kwargs,kde_kws=kwargs)
```

```python
# Box-cox
x_t,l=stats.boxcox(x) # l=lambda ,x_t= x transformed by box-cox
print(l)

## QQ plot:
stats.probplot(x_t,dist='norm',plot=plt)
plt.grid()
```
